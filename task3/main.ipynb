```python
def counting_len_way(sim_pred, sim_cur, pred_cost):
    global coord_simbols
    coords_sim_cur = coord_simbols[sim_cur]
    coords_sim_pred = coord_simbols[sim_pred]
    new_cost = [0 for _ in range(len(coords_sim_pred))]
    for ind, (x_pred, y_pred) in enumerate(coords_sim_pred):
        minimum = float('inf')
        for x_cur, y_cur in coords_sim_cur:
            print(coords_sim_pred, ind)
            cost = abs(x_cur - x_pred) + abs(y_cur - y_pred) + pred_cost[ind]
            if cost < minimum:
                minimum = cost
        new_cost[ind] = minimum
    return new_cost
```

```python
from collections import deque
import heapq
import sys

# ------------------------------
# ВВОД
# ------------------------------
n, m = map(int, input().split())
x0_, y0_ = map(int, input().split())
x0, y0 = x0_ - 1, y0_ - 1

grid = [input().strip() for _ in range(n)]
way = input().strip()               # порядок доставки
k = len(way)

# ------------------------------
# СОБИРАЕМ КООРДИНАТЫ ПО ТИПАМ
# ------------------------------
coords = {}
for i in range(n):
    for j in range(m):
        coords.setdefault(grid[i][j], []).append((i, j))

# ------------------------------
# ФУНКЦИИ BFS/ДЕЙКСТРА
# ------------------------------

# Мульти-источниковый BFS: расстояние до ближайшего типа "ch"
def bfs_from_all(ch):
    dist = [[10**15]*m for _ in range(n)]
    q = deque()
    for x, y in coords[ch]:
        dist[x][y] = 0
        q.append((x, y))
    while q:
        x, y = q.popleft()
        d = dist[x][y] + 1
        if x > 0 and dist[x-1][y] > d:
            dist[x-1][y] = d; q.append((x-1, y))
        if x+1 < n and dist[x+1][y] > d:
            dist[x+1][y] = d; q.append((x+1, y))
        if y > 0 and dist[x][y-1] > d:
            dist[x][y-1] = d; q.append((x, y-1))
        if y+1 < m and dist[x][y+1] > d:
            dist[x][y+1] = d; q.append((x, y+1))
    return dist

# Мульти-источниковая Дейкстра:
# стартовые вершины — ВСЕ клетки типа ch
# начальная цена = cost[ch_cell]
def dijkstra_from_all(cost_map, ch):
    dist = [[10**15]*m for _ in range(n)]
    pq = []

    # Кладём все клетки типа ch как источники
    for x, y in coords[ch]:
        start_cost = cost_map[x][y]
        dist[x][y] = start_cost
        heapq.heappush(pq, (start_cost, x, y))

    # Обычно все рёбра = 1, но нам позволяет Дейкстра с разными стартовыми весами
    while pq:
        cd, x, y = heapq.heappop(pq)
        if cd != dist[x][y]:
            continue
        nd = cd + 1
        if x > 0 and dist[x-1][y] > nd:
            dist[x-1][y] = nd; heapq.heappush(pq, (nd, x-1, y))
        if x+1 < n and dist[x+1][y] > nd:
            dist[x+1][y] = nd; heapq.heappush(pq, (nd, x+1, y))
        if y > 0 and dist[x][y-1] > nd:
            dist[x][y-1] = nd; heapq.heappush(pq, (nd, x, y-1))
        if y+1 < m and dist[x][y+1] > nd:
            dist[x][y+1] = nd; heapq.heappush(pq, (nd, x, y+1))
    return dist

# ------------------------------
# ОСНОВНОЙ АЛГОРИТМ
# ------------------------------

# С последней буквы: просто BFS до ближайшего нужного типа
last = way[-1]
cost = bfs_from_all(last)   # cost[x][y] = расстояние до ближайшего last

# Двигаемся назад по строке
for i in range(k-2, -1, -1):
    ch = way[i]
    # теперь хотим: cost_after_ch = min_{q тип ch+1} dist(p,q) + cost_after_{ch+1}[q]
    # это делает мульти-Дейкстра с начальными весами cost
    cost = dijkstra_from_all(cost, ch)

# ------------------------------
# ОТВЕТ: стоимость в точке старта
# ------------------------------
print(cost[x0][y0])
```
